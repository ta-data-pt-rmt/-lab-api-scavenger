# Challenge 1 : Fork languages

1. Obtain the full list of forks created from the main lab repo via Github API.

```python
# Specify the repository details
repo_owner = "ironhack-datalabs"
repo_name = "bcn-feb-2019"

# GitHub API URL for listing forks
api_url = f"https://api.github.com/repos/{repo_owner}/{repo_name}/forks"

# Send a GET request to the GitHub API
response = requests.get(api_url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Load the JSON data from the response
    data = response.json()

print(data)
```

```python
# Convert it into df :
df = pd.DataFrame(data)

df.head(3)
```

2. Loop the JSON response to find out the language attribute of each fork. Use an array to store the language attributes of each fork.

```python
#df.columns

unique_languages = df[df['language'].notna()]['language'].unique()

list(unique_languages)
```

```python
import json_normalize

# Send a GET request to the GitHub API
response = requests.get(url, headers=headers)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Load the JSON data from the response
    data_commits = response.json()

print(data_commits)
```

```python
df_commits = pd.DataFrame(data_commits)

df_commits.head()
```

```python
df_commits = pd.json_normalize(data_commits)
#normalize to convert nested or semi-structured data into a flat table-like structure. 
#This is especially useful when dealing with JSON objects that contain nested 
#dictionaries or arrays.
```

```python
df_commits
```

```python
# Convert 'commit.author.date' to datetime
df_commits['commit.author.date'] = pd.to_datetime(df_commits['commit.author.date'])
```

```python
#Filter 
df_commits[df_commits['commit.author.date'] > '2023-11-07']
```

```python
#Time delta method

from datetime import datetime, timedelta, timezone

# Calculate the date three weeks ago from today
three_weeks_ago = datetime.utcnow() - timedelta(weeks=3)
three_weeks_ago_utc = three_weeks_ago.replace(tzinfo=timezone.utc)

filtered_commits = df_commits[df_commits['commit.author.date'] >= three_weeks_ago_utc]

filtered_commits
```
